```python
# import libs do mod pyspark
from pyspark import SparkContext, SparkConf

# definir configuracao
conf = SparkConf().setAppName("WordCount")

# instanciar spark context
sc = SparkContext.getOrCreate()

# O conteúdo do arquivo é lido e carregado em um RDD (Resilient Distributed Dataset)
conteudo = sc.textFile("/home/jovyan/README.md")
# print(conteudo)

# filtrar linhas nao vazias, pra capturar as palavras
filter_lines = conteudo.filter(lambda x: len(x) > 0)
# print(filter_lines)

# função flatMap para dividir cada linha em palavras, usando um espaço em branco como delimitador.
words = filter_lines.flatMap(lambda x: x.split(' '))

# função map para mapear cada palavra e 'reducebykey' para somar o numero de ocorrências
word_count = words.map(lambda x: (x,1)).reduceByKey(lambda x, y: x + y).map(lambda x: (x[1], x[0])).sortByKey(False)

for word in word_count.collect():
    print(word)

# encerra o contexto spark
sc.stop()
```
